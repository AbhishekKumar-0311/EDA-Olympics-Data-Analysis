1. Groupby perform 3 important functions.
2. Split a dataframe into groups based on some criteria.

5. Groupby object is not a dataframe but a group of dataframes.

## Plot the number of medals won by US male and female athletes throughout the history of Olympics

```python
# Creating a dF of USA Athletes

usm = have[have.NOC == 'USA']
usm
```

```python
# Finding count of Medals won by each gender in each edition i.e, throughout history

usm.groupby(['Edition','Gender']).agg({'Medal' :'count'})
```

### View 0

```python
# Plotting the above with Bar-chart

usm.groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0).plot( figsize = (20,8));
```

### View 1

```python
# Plotting the above with Bar-chart

usm.groupby(['Edition','Gender']).agg({'Medal' :'count'}).plot(kind = 'bar', figsize = (25,8));
```

### View 2

```python
# Plotting the above result with Seaborn.
# Seaborn's countplot() has helped to directly plot medals count gender-wise (hue='Gender') throughout history(x='Edition').
# figure(figsize=()) - helps to customize the resultant figure size.

fig_dims = (25, 8)
plt.figure(figsize=fig_dims)
sbn.countplot(x= 'Edition', data= usm, hue= 'Gender');
```

## Plot the five athletes who have won the most gold medals over the history of Olympics. In case of tie, consider the number of silver medals and then bronze.

```python
# unstack() is used to make the medals count horizontal
# fill_value = 0 is to replace NaN with 0
# ascending = False to get the maximum on top
# figsize() - to provide custom size
# [['Gold','Silver','Bronze']] - to provide the order of Series

ag = have.groupby(['Athlete','Medal']).size().unstack('Medal', fill_value=0)
Medal_ordered = ag.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']] # The ['Gold','Silver','Bronze'] list series is included to order the ColumnNames
top5 = Medal_ordered.head(5)
top5.plot(kind='bar', figsize=fig_dims, title= 'Top 5 Gold winning Athletes', fontsize = 16);
```

## Seaborn with heatmaps

### Using the Olympic dataset, presenting a summary of the total medal won by participating country in 2008

```python
twoK8 = have[have.Edition == 2008]
smmr = twoK8.groupby(['NOC', 'Medal']).size().unstack('Medal', fill_value=0)
smm_sor = smmr.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']]
smm_sor
```

```python
sbn.heatmap(smm_sor);
```

Note :
1. The challenge with this plot is that the countries are on y axis, making is difficult to view. We can transpose this.

```python
smm_sor_trn = smm_sor.transpose()
smm_sor_trn
```

```python
plt.figure(figsize=(25,8))
sbn.heatmap(smm_sor_trn);
```

1. USA has highest no of Golds.
2. Australia has won Bronze more than anyone else.
3. Spain and Cuba has highest proportion of medals in Silver category.
